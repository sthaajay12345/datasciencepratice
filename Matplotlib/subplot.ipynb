```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random 
import matplotlib import style
```

```python
plt.subplot(2,2,1)
plt.pie([1])
plt.subplot(2,2,2)
plt.pie([1,2])
plt.subplot(2,2,3)
plt.pie([1,2,3])
plt.subplot(2,2,4)
plt.pie([1,2,3,4])
plt.show()
```

```python
plt.figure(figsize=(16,12))
plt.subplot(3,2,1)
a=np.arange(1,11)
b=a*2
c=[3,9,4,6,8,7,3,6,9,10]
d=a*4
plt.plot(a,b,'rD--',linewidth=3,label='b')
plt.plot(a,c,'g^:',linewidth=2,label='c')
plt.plot(a,d,'ko-.',linewidth=3,label='d')
plt.legend(loc='best')
plt.title('first graph',fontsize=16)
plt.xlabel('x_axis',fontsize=10)
plt.ylabel('y_axis',fontsize=14)
plt.grid()
plt.subplot(3,2,2)
a=np.random.randint(18,45,(100))
c=np.random.randint(15,45,(100))
b=[15,20,25,30,35,40,45]
plt.hist([a,c],color=["r","m"],bins=b,histtype='bar',rwidth=0.8,label=['ageof ml','age of al'])
plt.legend()
plt.subplot(3,2,3)
df=pd.read_csv('E:\\abc.csv',nrows=1000)
x=df['Rating']
y=df['Reviews']
plt.title("Google play store apps scatter plot")
plt.xlabel("Rating")
plt.ylabel('Review')
plt.scatter(x,y)
cmap='spectral'
plt.colorbar()
plt.subplot(3,2,4)
cls=['AI','DS','Python','ML','R']
cls1=[30,10,20,25,10]
cls2=[40,5,20,20,25]
cls3=[35,5,30,15,15]
a=np.arange(len(cls))
width=0.2
plt.bar(a,cls1,width,color='r',edgecolor='y',label='cls1')
plt.bar(a+width,cls2,width,color='g',label='cls2') # width represent move to next side of that bar
plt.bar(a+width+width,cls3,width,color='k',label='cls3')
plt.legend(loc='best')
plt.title("student of IT",fontsize=20)
plt.xlabel("course",fontsize=15)
plt.ylabel("student",fontsize=15)
plt.xticks(a+width,cls)
plt.subplot(3,2,5)
classes=['Python','R','Machine Learning','Artificial Internet','Data Science']
class1_students=[45,15,35,25,30]
a=[0,0,0,0,0.2]
wedgeprops={'linewidth':4,'width':1,'edgecolor':'k'}
plt.pie(class1_students,labels=classes,explode=a,autopct='%0.2f%%',radius=1.2,wedgeprops=wedgeprops)
plt.subplot(3,2,6,projection='polar',facecolor='k',frameon=True)
plt.show()
```
