```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
classes=['Python','R','Machine Learning','Artificial Internet','Data Science']
class1_students=[45,15,35,25,30]
plt.pie(class1_students)
plt.show()
```

```python
plt.pie(class1_students,labels=classes)
plt.show()
```

```python
classes=['Python','R','Machine Learning','Artificial Internet','Data Science']
class1_students=[45,15,35,25,30]
a=[0,0,0,0,0.2]
plt.pie(class1_students,labels=classes,explode=a)
plt.show()
```

```python
 classes=['Python','R','Machine Learning','Artificial Internet','Data Science']
class1_students=[45,15,35,25,30]
a=[0,0,0,0,0.2]
plt.pie(class1_students,labels=classes,explode=a,autopct='%0.2f%%',radius=1.9)
plt.show()
```

```python
 classes=['Python','R','Machine Learning','Artificial Internet','Data Science']
class1_students=[45,15,35,25,30]
a=[0,0,0,0,0.2]
wedgeprops={'linewidth':4,'width':1,'edgecolor':'k'}
plt.pie(class1_students,labels=classes,explode=a,autopct='%0.2f%%',radius=1.9,wedgeprops=wedgeprops)
plt.show()
```
