```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
cls=['AI','DS','Python','ML','R']
cls1=[30,10,20,25,10]
plt.bar(cls,cls1,width=0.5,align='center')
plt.show()
```

```python
cls=['AI','DS','Python','ML','R']
cls1=[30,10,20,25,10]
plt.bar(cls,cls1,align='edge')
plt.show()
```

```python
cls=['AI','DS','Python','ML','R']
cls1=[30,10,20,25,10]
plt.bar(cls,cls1,align='edge',color='r',edgecolor='k',linewidth =3,ecolor='y')
plt.show()
```

```python
# stacked bar graph
# in stack bar graph previous bar cls1 overlap by new bar cls2
```

```python
cls=['AI','DS','Python','ML','R']
cls1=[30,10,20,25,10]
cls2=[40,5,20,20,25]
plt.bar(cls,cls1,align='edge',color='r',linewidth =3,)
plt.bar(cls,cls2,align='edge',color='g',linewidth =3)
plt.title("student of IT",fontsize=20)
plt.xlabel("course",fontsize=15)
plt.ylabel("student",fontsize=15)
plt.show()
```

```python
# to show appropriate way of multiple bar graph
cls=['AI','DS','Python','ML','R']
cls1=[30,10,20,25,10]
cls2=[40,5,20,20,25]
cls3=[35,5,30,15,15]
a=np.arange(len(cls))
width=0.2
plt.bar(a,cls1,width,color='r',edgecolor='y',label='cls1')
plt.bar(a+width,cls2,width,color='g',label='cls2') # width represent move to next side of that bar
plt.bar(a+width+width,cls3,width,color='k',label='cls3')
plt.legend(loc='best')
plt.title("student of IT",fontsize=20)
plt.xlabel("course",fontsize=15)
plt.ylabel("student",fontsize=15)
plt.xticks(a+width,cls)
plt.show()
```
