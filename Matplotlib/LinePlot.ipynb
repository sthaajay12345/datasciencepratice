```python
days=np.arange(1,16)
days
```

```python
temperature=[24,34.4,44,32,33,25.5,31.1,28,43,43.3,30,32,36,37.4,40]
```

```python
plt.plot(days,temperature,"r^--")#here r =color,^=marker point,-- linestyle
plt.title('kathmandue temperature')
plt.xlabel('days')
plt.ylabel('temperature')
plt.show
```

```python
plt.plot(temperature,days)
plt.show()
```

```python
plt.plot(days,temperature,color='orange',linestyle=':',linewidth=3)
plt.title('my first graph')
plt.axis([0,30,0,50])
plt.xlabel('days')
plt.ylabel('temperature')
plt.show()
```

**Two line plot in same graph**

```python
a=np.arange(1,11)
y1=a*2
y2=a*3
```

```python
plt.plot(a,y1,color='orange',linestyle=':',linewidth=2)
plt.plot(a,y2,color='red',linestyle='-',linewidth=3)
plt.title("two line graph")
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.grid(True)
plt.show()

```

```python
plt.plot(a,y1,color='orange',linestyle=':',marker='o',linewidth=2)
plt.plot(a,y2,color='red',linestyle='-',marker='^',linewidth=3)
plt.title("two line graph")
plt.xlabel('x_axis')
plt.ylabel('y_axis')

plt.show()
```

```python
days =[1,2,3,4,5,6,7]
max_t=[50,41,52,48,49,46,47]
min_t=[43,42,44,33,35,45,32]
avg_t=[45,48,49,46,40,41,38]
 
```

```python
plt.plot(days,max_t,'ro--',label='max')
plt.plot(days,min_t,'b^:',label='min')
plt.plot(days,avg_t,'gD-.',label='avg')
plt.title('weather',fontsize=15)
plt.xlabel('days',fontsize=15)
plt.ylabel('temperature',fontsize=15)
plt.legend(loc='best')
plt.show()
```

```python
a=np.arange(1,11)
b=a*2
c=[3,9,4,6,8,7,3,6,9,10]
d=a*4
```

```python
plt.plot(a,b,'rD--',linewidth=3,label='b')
plt.plot(a,c,'g^:',linewidth=2,label='c')
plt.plot(a,d,'ko-.',linewidth=3,label='d')
plt.legend(loc='best')
plt.title('first graph',fontsize=16)
plt.xlabel('x_axis',fontsize=15)
plt.ylabel('y_axis',fontsize=14)
plt.grid()
plt.show()
```

```python
x=np.arange(1,11)
y1=x*2
y2=x*3
plt.plot(x,y1,'ro-.',linewidth=2)
plt.plot(x,y2,'ko-.',linewidth=2)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,'ro-.',linewidth=2)
plt.subplot(2,1,2)
plt.plot(x,y2,'k>:',linewidth=3)
plt.show()
```
