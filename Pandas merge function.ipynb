```python
import pandas as pd
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4,6],
                   'name':['a','b','c','d','e']})
df1
```

```python
df2=pd.DataFrame({'ID':[1,2,3,5],
                 'class':[6,7,8,10]})
df2
```

```python
pd.merge(df1,df2,on='ID')
```

```python
pd.merge(df1,df2,on='ID',how='left')
```

```python
pd.merge(df1,df2,on='ID',how='right')
```

```python
pd.merge(df1,df2,on='ID',how='inner')
```

```python
pd.merge(df1,df2,on='ID',how='outer')
```

```python
df3=pd.DataFrame({'ID':[1,2,3,4,6],
                   'name':['a','b','c','d','e']})
df3
```

```python
pd.merge(df1,df3,on="ID",suffixes=('_a','_b'))
```

```python
pd.merge(df1,df3,)
```
