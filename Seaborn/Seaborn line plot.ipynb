```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
days=np.arange(1,16)
temp=[24,40,44,32,33,25.5,31.1,28,43,43.3,30,32,36,37.4,40]
temp_df=pd.DataFrame({'days':days,'temp':temp})
sns.lineplot(x="days",y="temp",data=temp_df)
plt.show()
```
