```python
plt.figure(figsize=(16,9))
sns.heatmap(ab,annot=True)
```

```python
annot_kws={'fontsize':10,'fontstyle':'italic','color':'k','rotation':'vertical','verticalalignment':'center','backgroundcolor':'w'}
sns.heatmap(ab,annot=True,annot_kws=annot_kws,cmap='coolwarm')
```

```python
annot_kws={'fontsize':10, 
           'fontstyle':'italic',  
           'color':"k",
           'alpha':0.6, 
           'rotation':"vertical",
           'verticalalignment':'center',
           'backgroundcolor':'w'
           }
plt.figure(figsize=(16,9))
sns.heatmap(ab, annot = True, annot_kws= annot_kws,linewidth=4,linecolor=None)
```

```python
plt.figure(figsize=(16,9))
cbar_kws={'orientation':'horizontal','shrink':1,'extend':'max','extendfrac':0.1,'ticks':np.arange(0,21)}
sns.heatmap(ab,cbar_kws=cbar_kws)
```

```python
sns.heatmap(ab,xticklabels=False)
```

```python
a=sns.heatmap(ab)
a.set(title="heat map of global warming",
       xlabel='emmision',
       ylabel='country name')
sns.set(font_scale=1)
```

```python
plt.figure(figsize=(16,16))
a=sns.heatmap(ab.corr(),annot=True,linewidth=3)
a.tick_params(size=10,color='w',labelsize=10,labelcolor='w')
plt.title("heatmap of who is responsible for global waring",fontsize=15)
```
